```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
beta = 10
n = 15
k = 10
delta = 0.5
```

```python
B_vec = np.arange(0, 25, 0.01)
A_vec = np.arange(0, 25, 0.01)
A = beta*B_vec**n/(k**n + B_vec**n)/delta
B = beta*A_vec**n/(k**n + A_vec**n)/delta
plt.figure()
plt.plot(A, A_vec, label = "dA/dt = 0", color = "salmon")
plt.plot(B_vec, B, label = "dB/dt = 0", color = "aqua")
plt.xlabel('A')
plt.ylabel('B')
plt.legend()
plt.title("Phase Diagram")
plt.show()
```

```python
B_vec = np.arange(0, 25, 0.01)
A_vec = np.arange(0, 25, 0.01)
s_vec = np.arange(0.5, 19.5, 0.01)
A = beta*B_vec**n/(k**n + B_vec**n)/delta
B = beta*A_vec**n/(k**n + A_vec**n)/delta
s = -1 * s_vec + 19.9
plt.figure()
plt.plot(B, A_vec, label = "dA/dt", color = "salmon")
plt.plot(B_vec, A, label = "dB/dt", color = "aqua")
plt.plot(s_vec, s, '--', color = "orchid")
plt.xlabel('A')
plt.ylabel('B')
plt.legend(loc = "upper left")

x1vec = np.linspace(0, 25, 22)
x2vec = np.linspace(0, 25, 22)
X1, X2 = np.meshgrid(x1vec, x2vec)
Z1 = beta*X2**n/(k**n + X2**n) - delta*X1
Z2 = beta*X1**n/(k**n + X1**n) - delta*X2
Z1 = Z1 / np.sqrt(Z1**2+Z2**2)
Z2 = Z2 / np.sqrt(Z1**2+Z2**2)
plt.quiver(X2,X1,Z2,Z1,scale=25, color='lightgray', width=0.003)
plt.title("Phase Diagram with Slope Field")
plt.show()
```

```python
plt.figure()
B_vec = np.arange(0, 25, 0.01)
A_vec = np.arange(0, 25, 0.01)
A = beta*B_vec**n/(k**n + B_vec**n)/delta
B = beta*A_vec**n/(k**n + A_vec**n)/delta
plt.plot(B, A_vec, label = "dA/dt = 0", color = "salmon")
plt.plot(B_vec, A, label = "dB/dt = 0", color = "aqua")
for kb in range(10, 30, 2):
    A = beta*B_vec**n/(kb**n + B_vec**n)/delta
    B = beta*A_vec**n/(k**n + A_vec**n)/delta
    plt.plot(A, A_vec, ':', label = "dB/dt", color = "salmon")
B = beta*A_vec**n/(k**n + A_vec**n)/delta
plt.xlabel('A')
plt.ylabel('B')
plt.title("Phase Diagram for Increasing $k_b$")
plt.legend(["dA/dt","dB/dt"], loc = 'center right')
plt.show()
```
